#### 1) 문제 정의 (MD)

리스트에 n개의 항목이 들어있다. 이 리스트에서 "탐색키"를 가진 항목을 찾아라. 단, 리스트의 항목들은 정렬되어 있다.

#### 2) 알고리즘 설명  (MD)

이진 탐색 알고리즘은 반복구조나 순환구조로 모두 기술할 수 있다. 이진 탐색을 위한 실행 효율성을 위해서는 반복구조가 더 유리하다.

#### 4) 알고리즘 개요(입력 변수, 출력, 알고리즘 개요) (MD)

2 ~ 9행 : low와 high가 역전되지 않는 동안 같은 동작을 반복한다. 7행과 9행에서 low와 high만 잘 갱신해 주면 된다.

#### 5) 알고리즘 코드 (PY,C)

```python
def binary_search_iter(A, key, low, high) :
    while (low <= high) :
        mid = (low + high) // 2
        if key == A[mid]:
            return mid
        elif key > A[mid]:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

#### 6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def binary_search_iter(A, key, low, high) :
    while (low <= high) :
        mid = (low + high) // 2
        if key == A[mid]:
            return mid
        elif key > A[mid]:
            low = mid + 1
        else:
            high = mid - 1
    return -1

listA = [1, 3, 8, 13, 13, 16, 21, 26, 27, 30, 33, 36, 39, 41, 44, 49]
print("입력 리스트 = ", listA)
print("33 탐색(순환) --> ", binary_search_iter(listA, 33, 0, len(listA)-1))
print("33 탐색(반복) --> ", binary_search_iter(listA, 33, 0, len(listA)-1))
print("32 탐색(순환) --> ", binary_search_iter(listA, 32, 0, len(listA)-1))
print("32 탐색(반복) --> ", binary_search_iter(listA, 32, 0, len(listA)-1))
```

#### 8) 복잡도 분석

순차 탐색에서와 같이 이진 탐색에서도 입력의 구성에 따라 효율이 다르다. 최선의 경우는 테이블 중앙에 찾는 값이 있는 경우로 한 번의 비교만으로 알고리즘이 종료된다. 따라서 시간복잡도는 O(1)이다.
