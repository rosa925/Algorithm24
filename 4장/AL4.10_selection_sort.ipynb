#### 1) 문제 정의 (MD)

리스트에서 k번째로 작은 항목을 찾아라. 단, 리스트는 정렬되어 있지 않다.

#### 2) 알고리즘 설명  (MD)

정렬을 이용하여 k번째 작은 수를 찾는 알고리즘이다.

####  3) 손으로 푼 예제 (MD, 이미지 삽입)

#### 4) 알고리즘 개요(입력 변수, 출력, 알고리즘 개요) (MD)

2행 : 배열을 정렬한다.

3행 : 인덱스가 k-1인 항목을 반환한다.

#### 5) 알고리즘 코드 (PY,C)

```python
def kth_smallest_sort(A, k):
    A.sort()
    return A[k-1]
```

#### 6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def kth_smallest_sort(A, k):
    A.sort()
    return A[k-1]

test_list = [3, 1, 4, 1, 5, 9, 2, 6, 5]
k = 3
result = kth_smallest_sort(test_list, k)
print(f"{k}번째로 작은 원소는: {result}")
```
