## 1)문제 정의

모든 정점의 진입차수를 구하고, 진입차수가 0인 정점들을 리스트에 저장하고, 다음 리스트가 공백이 아닐때까지 진입차수가 0인 정점을 하나 꺼내 연결된 진입차수를 감소시킨다.

## 2)알고리즘 설명

1. 방향 그래프에서 진입차수가 0인 정점중에서 하나를 선택해 임의로 선택한다.
2. 선택된 정점을 삭제한다. 정점이 삭제되므로 이 정점에서 진출하는 모든 간선들도 삭제한다. 삭제되는 간선과 연결된 이웃 정점들의 진입 차수는 갱신되게 한다.
3. 이 과정을 모든 정점이 삭제될 때까지 반복한다. 정점이 삭제되는 순서가 위상 순서이다.

## 4)코드 개요

함수명: topological_sort
입력: mygraph = 
    "A": {"C", "D"},
    "B": {"D", "E"},
    "C": {"D", "F"},
    "D": {"F"},
    "E": {"F"},
    "F": {}
반환값: 반환값이 따로 없고 정렬된 요소를 출력합니다.
함수설명: 정점 진입차수를 계산하고, indeg[v]=0 초기화합니다. 진입차수가 0이면 vlist에 추가하고 진입차수가 0인 정점을 꺼내 출력합니다.
출력후 다음 정점의 진입차수를 감소하여 진입차수가 0인 리스트를 vlist에 추가합니다.


## 5)코드

```python
def topological_sort(graph):
    indeg = {}
    for v in graph:
        indeg[v] = 0
    for v in graph:
        for u in graph[v]:
            indeg[u] += 1

    vlist = []
    for v in graph:
        if indeg[v] == 0:
            vlist.append(v)

    sorted_elements = []
    while vlist:
        v = vlist.pop()
        sorted_elements.append(v)

        for u in graph[v]:
            indeg[u] -= 1
            if indeg[u] == 0:
                vlist.append(u)
```

## 6)테스트 코드

```python
def topological_sort(graph):
    indeg = {}
    for v in graph:
        indeg[v] = 0
    for v in graph:
        for u in graph[v]:
            indeg[u] += 1

    vlist = []
    for v in graph:
        if indeg[v] == 0:
            vlist.append(v)

    sorted_elements = []
    while vlist:
        v = vlist.pop()
        sorted_elements.append(v)

        for u in graph[v]:
            indeg[u] -= 1
            if indeg[u] == 0:
                vlist.append(u)

    print(' '.join(sorted_elements))

mygraph = {
    "A": {"B"},
    "B": {"C", "D"},
    "C": {"E",},
    "D": {"E"},
    "E": {"F"},
    "F": {}
}
print('topological_sort:')
topological_sort(mygraph)
print()
```
