### 01.문제정의
정렬되지 않은 리스트 A에서 탐색키를 가진 항목을 찾는 문제. 찾는 항목이 있으면 그 항목의 인덱스를 반환 하고 찾지 못하면 -1를 반환한다.

### 02.알고리즘 설명
정렬되지 않은 리스트 A에서 원하는 항목을 찾는 방법이다.

### 04. 코드개요
A=입력 리스트,key=탐색키
찾는 항목이 있으면 그 항목의 인덱스를 반환
찾지 모하면 -1를 반환

## 05.코드

```python
def sequential_search(A, key):
    for i in range(len(A)):
        if A[i] == key:
            return i
    return -1
```

### 06. 테스트 코드

```python
def sequential_search(A, key):
    for i in range(len(A)):
        if A[i] == key:
            return i
    return -1
result = sequential_search([1,2,3,4,5], 1)
print(result)
```
