### 4) 코드 개요(입력 변수, 출력, 함수 설명) (MD)

##### 1) 함수명 : def factorial_recur(n), 축소 정복 기법: 하향식 접근
##### 2) 입력 변수: n값을 받는다
##### 3) 반환값 : n값이 1이면 1을 반환하고, 1이 아니면 n값에서 -1을 하여 n값이 1이될 때까지 함수를 반복한다.

### 5) 코드 (PY,C)

```python
def factorual_recur(n) :
    if n == 1 :
        return 1
    else :
        return (n*factorual_recur(n-1))
```

### 6) 테스트 코드  (PY,C) , 입력을 바꾸어서

```python
def factorial_recur(n):
    if n == 1:
        return 1
    else:
        return n * factorial_recur(n - 1)

print(f"10!는 {factorial_recur(10)}")
```
